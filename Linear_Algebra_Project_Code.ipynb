# **Brightness Adjustment (Scalar Multiplication)**
---



```python
from IPython.display import display, Image
import numpy as np
from PIL import Image as PILImage
import matplotlib.pyplot as plt

# Load your image using a library like PIL
image_path = "TestingImage.JPG"
image = np.array(PILImage.open(image_path))

# Define a brightness factor
brightness_factor_increase = 1.5  # Adjust this value for brightening or darkening, e.g., 1.5 for brightening, 0.5 for darkening

# Apply the brightness adjustment
brightened_image = np.clip(image * brightness_factor_increase, 0, 255)  # Clip values to be within the valid pixel intensity range (0-255)
# Convert back to uint8 if necessary
brightened_image = np.uint8(brightened_image)

plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('on')

# Separating line
##plt.axvline(x=image.shape[1], color='black', linestyle='--', linewidth=2)

# Brightened Image
plt.subplot(1, 2, 2)
plt.imshow(brightened_image)
plt.title('Brightened/Darkened')
plt.axis('on')
plt.show()




# Adjust layout for better appearance
##plt.tight_layout()

# Show the plots

```

# **Matrix Multiplication**

---



```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Load the image
image_path = "TestingImage.JPG"
image = np.array(Image.open(image_path))

# Rotation angle in radians
theta = np.radians(45)

# Define the 2D rotation matrix
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)],
                            [np.sin(theta), np.cos(theta)]])

# Get image dimensions
height, width = image.shape[:2]

# Coordinates of the center of the image
center_x, center_y = width // 2, height // 2

# Create an empty array for the rotated image
rotated_image = np.zeros_like(image)

# Apply rotation to each pixel
for y in range(height):
    for x in range(width):
        # Shift coordinates to the center
        shifted_x, shifted_y = x - center_x, y - center_y
        # Apply rotation
        rotated_coords = np.dot(rotation_matrix, np.array([shifted_x, shifted_y]))
        # Shift coordinates back to the original position
        new_x, new_y = int(rotated_coords[0] + center_x), int(rotated_coords[1] + center_y)

        # Ensure the new coordinates are within the image boundaries
        if 0 <= new_x < width and 0 <= new_y < height:
            # Perform the rotation in the new image
            rotated_image[y, x] = image[new_y, new_x]

# Display the original and rotated images
plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(rotated_image, cmap='gray')
plt.title('Rotated Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

# **Inversion**

---



```python
import numpy as np
from PIL import Image

# Load the image
image_path = "TestingImage.JPG"
image = np.array(Image.open(image_path))

# Invert the image
inverted_image = 255 - image

# Display the original and inverted images
plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(inverted_image, cmap='gray')
plt.title('Inverted Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
